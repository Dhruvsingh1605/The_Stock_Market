```python
import yfinance as yf
import pandas as pd
import os

filename = "/media/dhruv/Local Disk/InternShala_Assignment_1/Stocks.csv"  # Ensure your CSV file is in the same directory as this script

if not os.path.exists(filename):
    print(f"Error: File '{filename}' not found. Please make sure the file is in the current directory.")
else:
    stocks = pd.read_csv(filename)

    stocks["Ticker"] = stocks["Ticker"] + ".NS"

    period = input("Enter the period (e.g., 1mo, 3mo, 6mo, 1y): ")
    investment_amount = float(input("Enter the total investment amount (in INR): "))

    result = pd.DataFrame()

    failed_tickers = []

    for ticker, weightage in zip(stocks["Ticker"], stocks["Weightage"]):
        print(f"Fetching data for {ticker}...")

        try:
            data = yf.download(ticker, period=period, interval="1d")

            if data.empty:
                print(f"No data found for {ticker}. Skipping...")
                failed_tickers.append(ticker)
                continue

            amount_to_invest = investment_amount * weightage
            data["Investment"] = amount_to_invest

            data["Shares"] = data["Investment"] / data["Close"]

            
            data["Ticker"] = ticker

            result = pd.concat([result, data[["Ticker", "Close", "Investment", "Shares"]]])

        except Exception as e:
            print(f"Failed to fetch data for {ticker}: {e}")
            failed_tickers.append(ticker)

    # if not result.empty:
    #     result.reset_index(inplace=True)
    #     output_filename = "investment_plan.xlsx"
    #     result.to_excel(output_filename, index=False)
    #     print(f"Investment plan saved to '{output_filename}'")
    # else:
    #     print("No valid data was fetched. Please check the ticker symbols or the Yahoo Finance API.")

    # if failed_tickers:
    #     print("The following tickers failed to fetch data:")
    #     print(", ".join(failed_tickers))
```
